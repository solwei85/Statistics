```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("athletes.csv")
df.head()
```

```python
df.describe()
```

```python
#como los numero nulos en altura y peso son pocos (vs el total), vamos a eliminarlos.
df_clean= df.dropna(subset=["height", "weight"])
df_clean.describe()
```

```python
#Calcula la moda, la mediana, la desviació estàndard i la mitjana aritmètica. 
#Voy a calcularlo sobre altura. En describe ya se calculó mediana(50%= 1.76), std (0.113111)
#y promedio (1.766942), solo falta calcular moda.
df_clean.mode()
```

El valor que más se repite en altura es 1.8

```python
#Calcular la correlacion entre 2 atributos. En este caso, entre altura y peso.
df_clean["height"].corr(df_clean["weight"])
```

```python
#Dado que la correlación es entre -1 y 1, 
#podemos decir que tiene correlación positiva relativamente alta.
#Esto también lo podemos graficar
title= "Correlación entre altura y peso"
plt.figure(figsize =(10, 10))
sns.regplot(x=df_clean["height"], y=df_clean["weight"], scatter_kws={'alpha':0.5}, y_jitter= 0.1, x_jitter=0.1)
plt.title(title, fontsize=20, color= "Blue")
plt.ylabel('peso')
plt.xlabel('altura')
plt.show()
```
